## Ajuste Basico de la Data

```python
import numpy as np
import pandas as pd 
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
sales_train = pd.read_csv(os.path.join('data', 'sales_train_v2.csv'))
shops = pd.read_csv(os.path.join('data', 'shops.csv'))
items = pd.read_csv(os.path.join('data', 'items.csv'))
item_cats = pd.read_csv(os.path.join('data', 'item_categories.csv'))

sales_train.name = 'sales_train'
shops.name = 'shops'
items.name = 'items'
item_cats.name = 'item_cats'
```

### visualizar de la data

```python
# Dimenciones de la data
dflist = [sales_train, shops, items, item_cats]
print('(numero de filas, numero de columnas)')
print('training set: ', sales_train.shape)
print('num of shops: ', shops.shape)
print('num of items: ',items.shape)
print('num of item categories: ',item_cats.shape)
```

```python
# Nombre de los atributos de cada data
print(sales_train.columns)
print(shops.columns)
print(items.columns)
print(item_cats.columns)
```

### Merge sales con el resto del datasets

```python
data=sales_train
data = data.merge(shops, how = 'left', on = 'shop_id')
data = data.merge(items, how = 'left', on = 'item_id')
data = data.merge(item_cats, how = 'left', on = 'item_category_id')
```

### Ajustar la variable Date

```python
data['date'] = pd.to_datetime(data['date'], format='%d.%m.%Y')
data['month_name'] = data['date'].dt.month
data['year'] = data['date'].dt.year
data['day'] = data['date'].dt.day
data['weekday'] = data['date'].dt.weekday
data['weekday_name'] = data['date'].dt.day_name()
data['month_name'] = data['date'].dt.month_name()
```
